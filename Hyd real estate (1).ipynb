```python
def remove_pps_outlier(df):
    """
    Remove outliers from a DataFrame based on the 'rent_amount' column.

    Parameters:
    df (DataFrame): Input pandas DataFrame containing the 'rent_amount' column.

    Returns:
    DataFrame: DataFrame without the outliers in the 'rent_amount' column.
    """

    df_out = pd.DataFrame()
    for key, subdf in df.groupby('locality'):
        m = np.mean(subdf.rent_amount)
        std = np.std(subdf.rent_amount)
        reduced_df = subdf[(subdf.rent_amount > (m - std)) & (subdf.rent_amount < (m + std))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
```

```python
filtered_df=remove_pps_outlier(filtered_df)
```

```python
# Create dummy variables for the 'locality' column in the DataFrame
dummy = pd.get_dummies(filtered_df.locality).astype(int)

# Display the first few rows of the resulting dummy variables DataFrame
dummy.head()
```

```python
# Concatenate the original DataFrame 'filtered_df' with the dummy variables DataFrame,
# dropping the 'Gachibowli' column from the dummy variables to avoid multicollinearity
concat_df = pd.concat([filtered_df, dummy.drop('Gachibowli', axis='columns')], axis='columns')
```

```python
# Dropping the 'locality' column from the DataFrame 'concat_df'
concat_df = concat_df.drop(columns='locality')
```

```python
# Creating features matrix X by dropping specific columns from 'concat_df'
X = concat_df.drop(['rent_amount', 'bathroom', 'floor_from _top', 'total amenities', 'deposit'], axis='columns')

# Creating the target variable y by selecting the 'rent_amount' column from 'concat_df'
y = concat_df['rent_amount']

# Displaying the first few rows of the features matrix X
X.head()
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets (70% train, 30% test) using 'train_test_split'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

# Initializing a Linear Regression model
lr = LinearRegression()

# Fitting the model using the training data
lr.fit(X_train, y_train)

# Calculating the R-squared score of the linear regression model on the test data
r2_score = lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score
from sklearn.linear_model import LinearRegression

# Defining a ShuffleSplit cross-validation with 5 splits and a test size of 20%
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

# Calculating cross-validation scores using Linear Regression model
cross_val_scores = cross_val_score(LinearRegression(), X, y, cv=cv)

# Displaying the cross-validation scores
print(cross_val_scores)
 
```

```python
from sklearn.linear_model import Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, ShuffleSplit
import pandas as pd

def find_best_model_using_gridsearchcv(X, y):
    # Dictionary defining the models and their hyperparameters for GridSearchCV
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False],
                'n_jobs': [1, 2, 4]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    
    # List to store model scores
    scores = []
    
    # Defining the cross-validation strategy
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    # Iterating over each model and its parameters
    for algo_name, config in algos.items():
        # GridSearchCV setup for each model
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)

        # Appending model's best score and parameters to the scores list
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    # Returning results in a DataFrame with columns for model, best score, and best parameters
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

# Calling the function and displaying results
find_best_model_using_gridsearchcv(X, y)
```

```python
def predict_price(locality, property_size, type_bhk):
    # Finding the index of the locality in the feature columns
    loc_index = np.where(X.columns == locality)[0][0]

    # Creating an array of zeros with the length of the feature columns
    x = np.zeros(len(X.columns))
    
    # Assigning the property size and type_bhk to the corresponding indices in the array
    x[0] = property_size
    x[1] = type_bhk
    
    # If the locality exists in the columns, set the corresponding locality index to 1
    if loc_index >= 0:
        x[loc_index] = 1

    # Predicting the price based on the trained Linear Regression model (lr)
    return lr.predict([x])[0]


```
